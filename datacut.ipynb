```python
import os
import pandas as pd

# Ensure the file path is correct

file = r"data\taylor\taylor.csv"

# Check if the file exists
if not os.path.exists(file):
    raise FileNotFoundError(f"File not found: {file}")


# Read the CSV file with the appropriate encoding
try:
    df_taylor = pd.read_csv(file, encoding='latin1')
except UnicodeDecodeError:
    df_taylor = pd.read_csv(file, encoding='iso-8859-1')

# Exclude rows where the 'Tour' column value matches any of the specified values
tours_to_exclude = ['Fearless_Tour', 'Speak_Now_World_Tour', 'The_Red_Tour', 'The_1989_World_Tour']
df_taylor = df_taylor[~df_taylor['Tour'].isin(tours_to_exclude)]


# Print the updated DataFrame
```

```python
df_taylor.reset_index(drop=True, inplace=True)
```

```python
df_taylor.insert(0, "main_attraction", "Taylor Swift")
```

```python
df_taylor.insert(0, "event_id", range(1, 54))
```

```python
import json
df_taylor
```

```python
json_documents = df_taylor.to_dict(orient='records')
total_files = len(json_documents)


output_folder = 'json_files'
os.makedirs(output_folder, exist_ok=True)

#Each JSON document to a separate file in the specified folder
for i, doc in enumerate(json_documents):
    file_name = f'document_{i + 1}.json'
    file_path = os.path.join(output_folder, file_name)
    with open(file_path, 'w') as f:
        json.dump(doc, f, indent=4)



print(f'{total_files} files have been created.')

```

#METALLICA

```python
metallica_file = r"data\metallica\Metallica_gigs.csv"

if not os.path.exists(metallica_file):
    raise FileNotFoundError(f"File not found: {metallica_file}")

df_metallica = pd.read_csv(metallica_file, sep=';')
```

```python
df_metallica.insert(1, "main_attraction", "Metallica")
df_metallica['Date'] = pd.to_datetime(df_metallica['Date'])
```

```python
df_metallica = df_metallica[df_metallica['Date'].dt.year >= 2018]
```

```python
df_metallica
```

```python
df_metallica['Song'] = df_metallica['Song'].astype(str)
df_metallica['Songs'] = df_metallica.groupby(['Date', 'Event', 'Location', 'Venue'])['Song'].transform(lambda x: ', '.join(x))
```

```python
df_metallica.drop_duplicates(subset=['Date', 'Location', 'Venue'], keep='first', inplace=True)
df_metallica.reset_index(drop=True, inplace=True)
```

```python
df_metallica = df_metallica.astype(str)
```

```python
metallica_culumns_to_exclude = ['Setlist_no', 'Song', 'Is_cover', 'Song_details', 'Song_profile_link']
df_metallica = df_metallica.drop(columns=metallica_culumns_to_exclude)
```

```python
total_files += 1
df_metallica.insert(0, "event_id", range(total_files, total_files + len(df_metallica)))
```

```python
df_metallica
```

```python
df_metallica.info()
```

```python
json_metallica = df_metallica.to_dict(orient='records')

output_folder = 'json_files'
os.makedirs(output_folder, exist_ok=True)
#Each JSON document to a separate file in the specified folder
for i, doc in enumerate(json_metallica):
    print(i)
    file_name = f'document_{i + total_files}.json'
    file_path = os.path.join(output_folder, file_name)
    with open(file_path, 'w') as f:
        json.dump(doc, f, indent=4)

total_files += len(df_metallica)-1


print(f'{total_files} files have been created.')
```
